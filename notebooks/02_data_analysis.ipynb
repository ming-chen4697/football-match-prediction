# Football Match Prediction - Data Analysis

This notebook demonstrates statistical analysis of football match data.

## 1. Load and Prepare Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load processed data
matches_df = pd.read_csv('data/processed/matches.csv')

print(f'Dataset shape: {matches_df.shape}')
print('\nFirst few rows:')
print(matches_df.head())
print('\nData types:')
print(matches_df.dtypes)
```

## 2. Basic Statistics

```python
print('Match Statistics:')
print(matches_df.describe())
```

## 3. Team Analysis

```python
from src.analysis.stats_analyzer import StatsAnalyzer

analyzer = StatsAnalyzer()
team_stats = analyzer.calculate_team_stats(matches_df)

print('Top 10 Teams by Points:')
print(team_stats[['team', 'matches', 'wins', 'draws', 'losses', 'goals_for', 'goals_against', 'points']].head(10))
```

## 4. Head-to-Head Analysis

```python
# Get first two teams from top 10
team1 = team_stats.iloc[0]['team']
team2 = team_stats.iloc[1]['team']

h2h_data = analyzer.head_to_head_analysis(matches_df, team1, team2)

print(f'H2H Analysis: {team1} vs {team2}')
print(h2h_data)
```

## 5. Form Analysis

```python
# Analyze form for top team
form_data = analyzer.form_analysis(matches_df, team1, last_n_matches=10)

print(f'Recent Form for {team1} (last 10 matches):')
print(form_data)
```
