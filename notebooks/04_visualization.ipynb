# Football Match Prediction - Visualization Demo

This notebook demonstrates data visualization capabilities.

## 1. Load Data and Create Visualizer

```python
import pandas as pd
from src.analysis.stats_analyzer import StatsAnalyzer
from src.visualization.charts import FootballVisualizer

# Load data
matches_df = pd.read_csv('data/processed/matches.csv')

# Create analyzer and visualizer
analyzer = StatsAnalyzer()
visualizer = FootballVisualizer()

# Calculate team stats
team_stats = analyzer.calculate_team_stats(matches_df)

print(f'Loaded {len(matches_df)} matches and {len(team_stats)} teams')
```

## 2. Team Comparison Chart

```python
# Plot top teams by points
fig_points = visualizer.plot_team_comparison(team_stats, 'points')
fig_points.show()

# Export to HTML
visualizer.export_html(fig_points, 'reports/top_teams_by_points.html')
```

## 3. Goal Distribution

```python
# Plot goal distribution
fig_goals = visualizer.plot_goal_distribution(matches_df)
fig_goals.show()

visualizer.export_html(fig_goals, 'reports/goal_distribution.html')
```

## 4. Goals For vs Against

```python
import plotly.graph_objects as go

# Create scatter plot
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=team_stats['goals_for'],
    y=team_stats['goals_against'],
    mode='markers+text',
    text=team_stats['team'],
    textposition='top center',
    marker=dict(
        size=team_stats['points']/3,
        color=team_stats['points'],
        colorscale='Viridis',
        showscale=True,
        colorbar=dict(title='Points')
    )
))

fig.update_layout(
    title='Goals For vs Against (bubble size = points)',
    xaxis_title='Goals For',
    yaxis_title='Goals Against',
    hovermode='closest',
    template='plotly_white'
)

fig.show()
```

## 5. Win Rate Analysis

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Calculate overall statistics
total_home_wins = len(matches_df[matches_df['home_goals'] > matches_df['away_goals']])
total_draws = len(matches_df[matches_df['home_goals'] == matches_df['away_goals']])
total_away_wins = len(matches_df[matches_df['home_goals'] < matches_df['away_goals']])

# Create pie chart
fig = go.Figure(data=[go.Pie(
    labels=['Home Wins', 'Draws', 'Away Wins'],
    values=[total_home_wins, total_draws, total_away_wins],
    marker=dict(colors=['#2ca02c', '#7f7f7f', '#d62728'])
)])

fig.update_layout(
    title='Overall Match Outcome Distribution',
    template='plotly_white'
)

fig.show()
```
