# Football Match Prediction - Data Fetching

This notebook demonstrates how to fetch football data from various sources.

## 1. Setup and Imports

```python
import pandas as pd
import numpy as np
from loguru import logger

# Import custom modules
from src.data_sources.football_json_loader import FootballJsonLoader
from src.data_sources.statsbomb_loader import StatsBombLoader
from src.preprocessing import DataCleaner

logger.info('Imports completed')
```

## 2. Fetch Data from football.json

```python
# Initialize loader
loader = FootballJsonLoader()

# Fetch matches for England Premier League
matches_df = loader.fetch_matches('en', 2023)
print(f'Loaded {len(matches_df)} matches')
print(matches_df.head())
```

## 3. Fetch Teams Data

```python
# Fetch teams
teams_df = loader.fetch_teams('en', 2023)
print(f'Loaded {len(teams_df)} teams')
print(teams_df.head())
```

## 4. Data Cleaning

```python
# Check for missing values
print('Missing values before cleaning:')
print(matches_df.isnull().sum())

# Clean data
matches_clean = DataCleaner.clean_pipeline(matches_df)
print(f'\nShape after cleaning: {matches_clean.shape}')
print('\nMissing values after cleaning:')
print(matches_clean.isnull().sum())
```

## 5. Save Processed Data

```python
import os

# Create directories if they don't exist
os.makedirs('data/processed', exist_ok=True)

# Save processed data
matches_clean.to_csv('data/processed/matches.csv', index=False)
if not teams_df.empty:
    teams_df.to_csv('data/processed/teams.csv', index=False)

print('Data saved successfully!')
```
